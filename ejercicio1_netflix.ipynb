### Identificación y visualización de outliers en 'release_year' y 'date_added'

```python
import matplotlib.pyplot as plt
# Identificación y visualización de outliers en 'release_year'
plt.figure(figsize=(8, 3))
plt.boxplot(data_netflix['release_year'].dropna(), vert=False)
plt.title('Boxplot de release_year')
plt.xlabel('Año de lanzamiento')
plt.show()
```

```python
# Continuar con la identificación de outliers en 'release_year'
Q1 = data_netflix['release_year'].quantile(0.25)
print(f'Q1={Q1}')
Q3 = data_netflix['release_year'].quantile(0.75)
print(f'Q3={Q3}')
IQR = Q3 - Q1

# Identificar límites para detectar outliers
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# Filtrar outliers
outliers = data_netflix[(data_netflix['release_year'] < lower_bound) | (data_netflix['release_year'] > upper_bound)].release_year

# Mostrar resultados
outliers_summary = outliers.describe()

outliers_summary
```

```python
# Convertir la columna 'date_added' a timestamps
data_netflix['date_added_timestamp'] = data_netflix['date_added'].apply(lambda x: x.timestamp() if pd.notnull(x) else None)

# Crear el boxplot
plt.figure(figsize=(8, 3))
plt.boxplot(data_netflix['date_added_timestamp'].dropna(), vert=False)
plt.title('Boxplot de date_added')
plt.xlabel('Fecha de incorporación')

# Convertir los valores de timestamp a datetime para las etiquetas
ticks = plt.gca().get_xticks()
plt.gca().set_xticklabels([pd.to_datetime(t, unit='s').strftime('%Y-%m-%d') for t in ticks])
plt.show()
```

```python
# Continuar con la identificación de outliers en 'release_year'
Q1 = data_netflix['date_added'].quantile(0.25)
print(f'Q1={Q1}')
Q3 = data_netflix['date_added'].quantile(0.75)
print(f'Q3={Q3}')
IQR = Q3 - Q1

# Identificar límites para detectar outliers
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# Filtrar outliers
outliers2 = data_netflix[(data_netflix['date_added'] < lower_bound) | (data_netflix['date_added'] > upper_bound)].date_added

# Mostrar resultados
outliers_summary2 = outliers2.describe()

outliers_summary2
```
